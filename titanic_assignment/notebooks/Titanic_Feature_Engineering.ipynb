## 1. Setup and Data Loading

```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
base = Path('..').resolve()
data_dir = base / 'data'
outputs_dir = base / 'data'
train = pd.read_csv(data_dir / 'train.csv')
test = pd.read_csv(data_dir / 'test.csv') if (data_dir / 'test.csv').exists() else None
train.head()
```

## 2. Data Cleaning Decisions

```python
train.isna().sum().sort_values(ascending=False)
```

```python
# Run cleaning script
!python ../scripts/data_cleaning.py
```

```python
cleaned = pd.read_csv(outputs_dir / 'train_cleaned.csv')
cleaned.head()
```

### Outlier Visualization (Before/After)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.boxplot(y=train['Fare'], ax=axes[0])
axes[0].set_title('Fare Before Cleaning')
sns.boxplot(y=cleaned['Fare'], ax=axes[1])
axes[1].set_title('Fare After Cleaning')
plt.tight_layout()
```
